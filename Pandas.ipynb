# Concatenate Data Frames

```python
import pandas as pd
indian_weathers = pd.DataFrame({
    "City":["HYD","Banglore","Mumbai"],
    "temperature":[27,28,29],
    "Humidity":[68,65,61]
})
indian_weathers
```

```python
us_weathers = pd.DataFrame({
    "City":["okla","nyc","texas"],
    "temperature":[21,22,23],
    "Humidity":[78,85,91]
})
us_weathers
```

```python
df = pd.concat([indian_weathers, us_weathers], ignore_index=True)
df
```

```python
df = pd.concat([indian_weathers, us_weathers], axis=1, )
df
```

# Merge DataFrames | SQL Joins

```python
temperature_df = pd.DataFrame({
    "City":['Delhi','Banglore',"Hyderabad"],
    "Temperature":[41,42,39]
})
temperature_df
```

```python
humidity_df = pd.DataFrame({
    "City":["Delhi","Banglore","Kolkatta","Pune"],
    'Humidity':[87,58,54,21]
})

humidity_df
```

```python
#Merge two DataFrames without mention INDEXING
df = pd.merge(temperature_df,humidity_df,on='City')
df
```

```python
#OUTER_JOIN
df = pd.merge(temperature_df,humidity_df,on='City',how='outer')
df
```
